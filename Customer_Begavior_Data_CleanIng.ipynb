```python
import pandas as pd 
df = pd.read_csv("/Users/apple/Documents/customer_shopping_behavior Project/customer_shopping_behavior.csv")
# Here we can read file
df.head()
```

```python
df.info()
#Used to get a quick summary of your dataset.
```

```python
df.describe(include = 'all')
# It's use to get statistical summary of numerical columns in dataset and we use (include = 'all') for get all data like
# Numerical as well as object columns
```

```python
df.isnull().sum()
# Here we use this to check missing/null values 
```

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna (x.median()))
# This is use to fill missing/NaN values using group-wise median
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
# convert all columns name in lowerr case
df.columns
```

```python
df.columns = df.columns.str.replace(' ','_')
# fill all spaces with underscore(-)
df.columns
```

```python
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
#here we do rename column name
df
```

```python
# create a column age-group
values = ['Young Adult','Adult','Middle-age','Senior']
df['age_group'] =  pd.qcut(df['age'],q=4, labels = values)
# here we create a column and divide age column into 4 equal groups and assign meaningful labels to taht age_group
df[['age','age_group']].head()
```

```python
# create column purchase_ferquency_days

ferquency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Bi-Weekly' : 14,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Annually' : 365,
    'Every 3 Months' : 90
}

df['purchase_ferquency_days'] = df['frequency_of_purchases'].map(ferquency_mapping)
df[['purchase_ferquency_days','frequency_of_purchases']].head(10)
# here we convert text frequency into number of days using dictionary mapping (using map function)
```

```python
df[['discount_applied','promo_code_used']].head(10)
# Both columns have same values
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
# here we check both columns are same or not
```

```python
df = df.drop('promo_code_used',axis=1)
# here we drop/delete the promo_code_used column
```
